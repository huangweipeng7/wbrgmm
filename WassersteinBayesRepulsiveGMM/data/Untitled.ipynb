```python
import pandas as pd
import sklearn as sk
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('joensuu.csv', sep=' ')
```

```python
dp = sk.mixture.BayesianGaussianMixture(
    n_components=5,
    init_params='random',
    weight_concentration_prior=1,
    weight_concentration_prior_type='dirichlet_process'
)
```

```python
cls = dp.fit_predict(df.values)
```

```python
sns.scatterplot(data=df, x='x', y='y', c=cls, legend='outer')
```
