```julia
using CSV
using DataFrames
using Distributions
using LinearAlgebra
using Random
using Plots, StatsPlots
```

```julia
mu1, sig1 = [0., 0], Matrix([[1.  0]; [0 10]])
mu2, sig2 = [0., 0], Matrix([[2. -4.]; [-4 10]])
mu3, sig3 = [0., 0], Matrix([[2.  4.]; [ 4 10]])

G1 = MvNormal(mu1, sig1)
G2 = MvNormal(mu2, sig2)
G3 = MvNormal(mu3, sig3)

Random.seed!(200)
a = rand(G1, 100)
b = rand(G2, 100)
c = rand(G3, 100)
```

```julia
z = hcat(a, b, c)'
print(size(z))
C = vcat(fill(1, 100), fill(2, 100), fill(3, 100))
scatter(z[:, 1], z[:, 2], color=C, markersize=3, xlim=(-10, 10), ylim=(-10, 10))
```

```julia
df = DataFrame(:x => z[:, 1], :y => z[:, 2], :C => C)
CSV.write("/home/weipeng/Workspace/projects/wbrgmm/WassersteinBayesRepulsiveGMM/data/sim_data0.csv", df)
```
