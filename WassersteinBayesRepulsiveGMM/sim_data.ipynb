```python
import scipy
np.random.seed(11)

m1 = np.array([0, 0])
m2 = np.array([0, 0]) 
sig1 = np.array([[100, 0], [0, 1]])
sig2 = np.array([[1, 0], [0, 100]])
sig3 = [[4, 1], [1, 4]]

K = 1
# ww = scipy.stats.dirichlet(np.ones(K)).rvs(1)
ww = [1/4, 1/4, 1/2]
nn = scipy.stats.multinomial.rvs(200, ww)
print(nn)

x1 = np.random.multivariate_normal(m1, sig1, nn[0])
x2 = np.random.multivariate_normal(m2, sig2, nn[1]) 
data = [x1, x2]

print(ww)
for n in nn[2:]: 
    m = np.random.multivariate_normal([0, 0], [[100, 0], [0, 100]])
    data.append(np.random.multivariate_normal(m, sig3, n))

z = np.vstack(data)

df = pd.DataFrame({'x': z[:, 0], 'y': z[:, 1]})
sns.scatterplot(df, x='x', y='y', alpha=0.5)
df.to_csv(
    '~/Workspace/projects/wbrgmm/WassersteinBayesRepulsiveGMM/data/sim_data_new_1.csv',
    index=False
)
```

```python
from scipy.stats import invwishart

np.random.seed(110)

m1 = np.array([0, 0])
m2 = np.array([0, 0]) 
sig1 = np.array([[100, 0], [0, 1]])
sig2 = np.array([[1, 0], [0, 100]])
sig3 = [[10, 8], [8, 10]]

K = 4
# ww = scipy.stats.dirichlet(np.ones(K)).rvs(1)
ww = [1/12, 1/12, 1/6, 1/6, 1/6, 1/6, 1/6]
nn = scipy.stats.multinomial.rvs(500, ww)
print(nn)

x1 = np.random.multivariate_normal(m1, sig1, nn[0])
x2 = np.random.multivariate_normal(m2, sig2, nn[1]) 
data = [x1, x2]
C = [0] * nn[0] + [1] * nn[1]
gauss = [(m1, sig1), (m2, sig2)]

print(ww)
for i, n in enumerate(nn[2:]): 
    m = np.random.multivariate_normal([0, 0], [[100, 0], [0, 100]])
    sig = invwishart.rvs(2, sig3, 1)
    C += [i+2] * n
    gauss.append((m, sig))
    data.append(np.random.multivariate_normal(m, sig3, n))

z = np.vstack(data)

df = pd.DataFrame({'x': z[:, 0], 'y': z[:, 1], 'C': C})
sns.scatterplot(df, x='x', y='y', alpha=0.5)
df.to_csv(
    '~/Workspace/projects/wbrgmm/WassersteinBayesRepulsiveGMM/data/sim_data_new_6.csv',
    index=False
)
```
